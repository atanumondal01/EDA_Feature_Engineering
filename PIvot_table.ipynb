```python
import pandas as pd
import seaborn as sns
```

```python
df=sns.load_dataset("flights")
```

```python
df.head()
```

```python
x=df.pivot_table(index="year",columns="month",values="passengers",aggfunc="sum")
x
```

```python
sns.heatmap(x,annot=True,linewidths=0.5,fmt="d")#d is integer
```

```python
sns.heatmap(x,annot=True,linewidths=0.5,fmt="d",cmap="viridis")
#cap can be summer ,winter
```
